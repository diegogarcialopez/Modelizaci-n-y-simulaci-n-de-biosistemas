**Finding the maximum value of f(x)**

f'(x) = 0 => Maximum value of f(x)

f'(x) = 2rx - 3rx² ;

    2rx - 3rx² = 0 ; x = 0

    2r - 3rx = 0 ; x = 2/3 
                     
f(2/3) = 4r/27

With this information we can assume that: 0 < r < 27/4

**Finding fixed points**

x = r(1-x)x² ; rx² - rx + 1 = 0 

x = (r +- sqrt(r² -4))/2r


**Estimating the stability of the points**

For both points, and following the stability condition (f'(x)<1): 

-2 < r < 2

**Plotting the dynamics of the system**

```python
import matplotlib.pyplot as plt

def plotgraph(x,r):
    x_list = []
    for _ in range (100):
        x_list.append(x)
        y = r*(1-x)*x**2
        x = y
    iterations = list(range(0, 100))
    plt.plot(iterations, x_list)
    plt.title('x_0 =' + str(x_list[0]) + '; r =' + str(r))
    plt.xlabel("n")
    plt.ylabel("x");
```

```python
plot1 = plotgraph(0.75, 4.5)
```

```python
plot2 = plotgraph(0.75, 5)
```

```python
plot3 = plotgraph(0.75, 5.5)
```

It is a period 2 graph!

```python
plot4 = plotgraph(0.75, 5.8)
```

It is a period 4 graph!

```python
plot5 = plotgraph(0.75, 6)
```

It is a period 16 graph!

```python
plot6 = plotgraph(0.75, 6.5)
```

It is chaos!

**Plotting the bifurcation diagram**

```python
import matplotlib.pyplot as plt
import numpy as np

rate = np.linspace(4,6.5,10000)
r_list = []
x_list = []

for r in rate:
    r_list.append(r)
    x = np.random.random()
    for n in range(100):
        x = r*(1-x)*x**2
    for n in range(100):
        x = r*(1-x)*x**2
    x_list.append(x)
plt.plot(r_list,x_list, ls = '', marker = ',')
plt.show()
```
