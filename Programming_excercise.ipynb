Write a function y = f(r,x) that takes two numbers (r and x) as input, calculates y = rx(1-x), and returns the value of y.

```python
r = 3
x = 2
y = r*x*(1-x)
print(y)
```

Write a script or function that does the following: 

(i) it takes two numbers (x_0 and r) as input; 

(ii) then, it iteratively calls function y = f(r,x) in a loop, such that the value of y returned in each iteration is used as the new value of x for the next iteration. The loop should include 100 iterations, starting with x = x_0. 

(iii) At each iteration, record the value of x. 

(iv) After the last iteration, plot the series of values of x (the values of x should go in the vertical axis and the iteration number in the horizontal axis). 

If you find plotting the data challenging, you can just save the values of x in a text file and plot them with excel or any plotting software (e.g. xmgrace).

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

def plotgraph(x,r):
    x_list = []
    for _ in range (100):
        x_list.append(x)
        y = r*x*(1-x)
        x = y
    iterations = list(range(0, 100))
    plt.plot(iterations, x_list)
    plt.title('x_0 =' + str(x_list[0]) + '; r =' + str(r))
    plt.xlabel("n")
    plt.ylabel("x");
```

```python
plot1 = plotgraph(0.5, 2.7)
pp = PdfPages('Programming_excercise_Plots.pdf')
pp.savefig(plot1)
```
